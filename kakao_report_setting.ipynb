# 카카오 캠페인별 보고서 한 파일에 정리하는 함수
- 목적 : 카카오 광고 보고서는 한번에 다운 받을 수 없음
- 캠페인별 다운만 가능하므로 약 20개의 캠페인별 보고서를 한 파일에 합치기 번거로움
- 다운 받으면 하나로 합쳐주는 함수 생성

```python
#파일 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
import os

sns.set()

import missingno as msno
import warnings
warnings.filterwarnings("ignore")

from google.colab import files
uploaded = files.upload()
```

```python
#카카오 광고 성과 보고서 합치는 함수
def set_df():
  path = '/content/'
  file_list = os.listdir(path)
  file_list_py = [file for file in file_list if file.endswith('.csv')]
  file_list_py.sort()
  
  df = pd.DataFrame()
  for i in file_list_py:
    data = pd.read_csv(path + i)
    df = pd.concat([df,data])
    
  df = df.reset_index(drop = True)
  df.to_excel('kakao_0214.xlsx', index=False)
  return df
```
